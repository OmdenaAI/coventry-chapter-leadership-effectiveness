### Imports

```python
import os
from dotenv import load_dotenv
import tweepy
import pandas as pd
```

### Read Credentials

```python
load_dotenv()
```

```python
consumer_key = os.getenv('API_KEY')
consumer_secret = os.getenv('API_KEY_SECRET')
access_token = os.getenv('ACCESS_TOKEN')
access_token_secret = os.getenv('ACCESS_TOKEN_SECRET')
```

### API Connection

```python
auth = tweepy.OAuth1UserHandler(
   consumer_key, consumer_secret, access_token, access_token_secret
)
```

```python
api = tweepy.API(auth, wait_on_rate_limit=True)
```

### Data fetch
Empty set and list of keywords provided

```python
usernames = set()
```

```python
keywords = ['CEO', 'vice president', 'president',
                      'chief', 'founder', 'co funder', 'CTO', 'Congress Women', 'Congress men',
                      'senator', 'MP', 'parliament', 'head', 'senior', 'Activist', 'creator', 'board member',
                      'Chairman', 'VP']
```

Loop below iterates through keywords and saves accounts that have 100k and above followers into usernames

```python
for word in keywords:
    cursor = tweepy.Cursor(api.search_users,q=f"{word} lang:en",count=20)
    for user in cursor.items(1000):
        followers = user.followers_count
        if followers > 99999:                       
            print(user.screen_name)
            usernames.add(user.screen_name) 
```

```python
len(usernames)
```

```python
pool = list(usernames)
```

Empty pandas dataframe to fill

```python
df = pd.DataFrame(columns=["username","bio","profile_URL","website"])
```

Below loop iterates through list of usernames and saves their data into dataframe if their language is english

```python
count = 0
for item in pool:
    handle = item
    user = api.get_user(screen_name=handle)
    try:
        if user.status.lang == 'en':
            try:
                df = pd.concat([df, pd.DataFrame.from_records([{'username':user.screen_name,
                                                                'bio':user.description,
                                                                "profile_URL":'https://twitter.com/'+handle,
                                                                "website":user.entities['url']['urls'][0]['display_url']
                                                               }])
                               ], ignore_index=True)
            except:
                df = pd.concat([df, pd.DataFrame.from_records([{'username':user.screen_name,
                                                                'bio':user.description,
                                                                "profile_URL":'https://twitter.com/'+handle
                                                               }])
                               ], ignore_index=True)
            print(count)
            count+=1
    except:
        pass
```
